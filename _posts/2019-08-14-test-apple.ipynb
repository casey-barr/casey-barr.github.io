### Introduction 

As my first *official* blog post, I thought that I'd take a look and explore some data personal to me; my own Apple Health data, collected primarily from my two most frequently used Apple Devices, my iPhone and my Apple Watch.

To start, there'll be a brief write-up on how to get the data collected from Apple's Heatlh app, and then I'll move on to doing some EDA on that data in a Jupter notebook, as well as a machine learning model that I wrote in Tensorflow to see if my habits can be predicted. 

### Data Collection

I will be using [this guide](http://www.markwk.com/data-analysis-for-apple-health.html), written by [Mark Koester](http://www.markwk.com/), to extract and format the raw data stored on my Apple Devices.

We can begin with opening the Health app, selecting the User Icon in the top right corner of the screen, and then selecting the 'Export Health Data' option at the bottom of the page.

The Health app then collects all of the data stored on your device, and will provide you with a file named `export.zip` after a few moments of processing. You can then choose to share it with yourself through any of Apple's filesharing methods (email, Airdrop, etc.). I chose to email my data to myself, and downloaded the zip to my local computer.
